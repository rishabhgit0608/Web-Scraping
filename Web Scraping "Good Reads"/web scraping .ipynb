```python
from bs4 import BeautifulSoup
import requests
```

```python
BASE_URL="https://www.goodreads.com/quotes?page=1"
response = requests.get(BASE_URL)
```

```python
soup =BeautifulSoup(response.text,'html.parser')
soup.text
soup.prettify
```

```python
quotes=[]
```

```python
def parse_html(html_blobs):
    soup= BeautifulSoup(html_blobs,'html.parser')
    div_quotes= soup.find('div',attrs = {'class':'quotes'})
    for quote in div_quotes.select('div.quote') :
        quote_text = quote.select_one('div.quoteText')
        quotefooter= quote.select_one('div.quoteFooter')
        tags  =  quotefooter.find('div',attrs={'class':['greyText','smallText','left',]})
        
        tag= []
        if tags:
            anchor_tags = tags.find_all('a')
            anchor_likes= quotefooter.find('a',attrs = {'class':'smallText'})
            tag = [a.text for a in anchor_tags]
        quote = quote_text.text.strip().split('―')
        quote,author = quote[0].strip()[1:-1],quote[1].strip()
        author = author.split(',')[0]
        likes = int(anchor_likes.text.replace(' likes',''))
        print(quote,author,tags,likes,sep='\n',end = '\n\n')
        quotes.append({'author':author, 'quote':quote , 'tags':tag ,'likes':likes})
```

```python
def make_request():
    BASE_URL="https://www.goodreads.com/quotes?page="
    pages=(int)(input("Enter the number of page"))
    for page in range(pages):
        response=requests.get(BASE_URL+(str)(page+1))
        parse_html(response.text)  
        
make_request()
```

```python
import csv 

with open('quotes.csv', 'w') as f : 
    writer = csv.DictWriter(f,fieldnames=['quote','author','tags','likes'])
    writer.writeheader()
    for q in quotes:
        writer.writerow(q)
```
